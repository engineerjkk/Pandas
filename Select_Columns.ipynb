# NA는 비교가 불가합니다.  
NA 값은 NA 끼리도 비교하지 못합니다.  

```python
print(np.nan !=np.nan)
print(np.nan>1)
print(np.nan<1)
print(np.nan !=1)
```

오직 같지 않다 ! 만 참이 나오고 나머지가 거짓이 나옵니다.  
그러니, S나 DF를 비교할 때에도 null 값을 처리하지 않으면 문제가 됩니다. !  
아래를 보죠  

```python
import pandas as pd
a = pd.Series([1, 2, 3])
b = pd.Series([1, 2, 3])
c = pd.Series([1, np.nan, 3])
d = pd.Series([1, np.nan, 3])
```

```python
#무사통과하는 a == b
print(a == b)
print()
# 1번 인덱스를 잘 보세요.
print(c == d)
```

```python
#무사통과하는 a == b
print(a == b)
print()
# 1번 인덱스를 잘 보세요.
print(c == d)
```

a와 b가 동일한 요소를 가지고 있나요? 하면 NaN이어도 True입니다.

```python
print(a.equals(b))
print(c.equals(d))
```
