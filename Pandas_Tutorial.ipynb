# 열과 행의 선택
DF는 표 형태이기 때문에 열 혹은 행을 기준으로 데이터를 선택할 수 있습니다.
기본적으로 numpy의 array을 선택하는 것과 동일하게 데이터를 있습니다. 
'A'라는 열이 첫 열로 df에 있을때
* df.A
* df['A']
* df.loc[:,'A'] #행에대해 : 을 놓으면 모든 행을 의미한다.
* df.iloc[:,0] #0은 컬럼의 위치를 의미한다.       

1번과 2번 방법은 직관적입니다.   
3번 loc는 레이블 이름 기반으로 데이터에 접근을 합니다. 이름표를 보고 갑니다.   
4번 iloc는 위치 기반으로 데이터에 접근을 합니다. 데이터의 위치번호를 보고 접근합니다.   
그리고, 두 개의 차원을 표시할 때 :는 모든 것을 의미합니다. 즉, 모든 행을 선택하는데 열 이름이 'A'인 것입니다.   
movies의 director_name 칼럼에 접근해 보겠습니다.

```python
movies.loc[:,'director_name']
```

```python
movies.iloc[:,1]
```

```python
type(movies.director_name)
```

여러 방법이 있지만 가장 타이핑이 적은 방법은 .을 쓰는 것입니다.
게다가 jupyterlab에서는 자동완성도 지원해주니 편합니다.

```python
movies['total_likes']=(movies.actor_1_facebook_likes+
                      movies.actor_2_facebook_likes+
                      movies.actor_3_facebook_likes+
                      movies.director_facebook_likes)
```

동일한 결과를 sum 함수를 이용하여서도 만들 수 있습니다.  
axis=1은 컬럼 방향으로의 합을 의미합니다.  
기본값은 0이기에 1을 선택해줍시다. 우리는 각 영화들의 합을 원합니다.

```python
cols=['actor_1_facebook_likes',
                      'actor_2_facebook_likes',
                      'actor_3_facebook_likes',
                      'director_facebook_likes']
sum_col=movies.loc[:,cols].sum(axis=1)
sum_col
```

movies df는 상당히 크기 때문에, facebook 이 컬럼에 들어가 있는 것만 선택해서 새로운 df를 만들어 보겠습니다.  
movies df에 전체 컬럼이름 중에서 facebook이 들어가는지 확인하는 것입니다. 그럼 facebook 키워드가 있는 컬럼만 볼수 있는것입니다.

```python
movies_fb=movies[[
    c
    for c in movies.columns
    if 'facebook' in c
]]
```

```python
[
    c
    for c in movies.columns
    if 'facebook' in c
]
```

```python
movies_fb #6개의 컬럼만 선택한 결과를 본다.
```

각 영화마다 facebook_likes의 총합을 구할 수 있다.

```python
movies_fb.sum(axis=1) 
```

assign 함수는 total_like라는 컬럼을 만들어줍니다. 얘 값은 무엇으로 해줘! 라는 시리즈를 넣습니다. 

```python
movies_fb=movies_fb.assign(total_likes=movies_fb.sum(axis=1))
movies_fb.head()
```
